## Removing unwanted text

```python
import re

# Define a function to remove unwanted text
def clean_text(text):
  """
  Cleans the input text by removing header information and other unwanted parts.
  """
  text = re.sub(r"Xref:.+\n", "", text) # remove Xref line
  text = re.sub(r"Path:.+\n", "", text) # remove Path line
  text = re.sub(r"From:.+\n", "", text) # remove From line
  text = re.sub(r"Newsgroups:.+\n", "", text) # remove Newsgroups line
  text = re.sub(r"Subject:.+\n", "", text) # remove Subject line
  text = re.sub(r"Date:.+\n", "", text) # remove Date line
  text = re.sub(r"Organization:.+\n", "", text) # remove Organization line
  text = re.sub(r"Lines:.+\n", "", text) # remove Lines line
  text = re.sub(r"Message-ID:.+\n", "", text) # remove Message-ID line
  text = re.sub(r"References:.+\n", "", text) # remove References line
  text = re.sub(r"NNTP-Posting-Host:.+\n", "", text) # remove NNTP-Posting-Host line
  text = re.sub(r"In-reply-to:.+\n", "", text) # remove In-reply-to line
  text = re.sub(r"Distribution:.+\n", "", text) # remove Distribution line
  text = re.sub(r"Sender:.+\n", "", text) # remove Sender line

  # remove multiple newlines
  text = re.sub(r"\n+", "\n", text)

  text = text.strip()

  return text


data = blogs['Data'].apply(clean_text)
```

```python
data
```

```python
def clean_text1(text):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = re.sub("[0-9" "]+"," ",text)
    text = re.sub('[‘’“”…]', '', text)
    return text

data = data.apply(lambda x: clean_text1(x))
```

## N-Gram

```python
#Bi-gram
def get_top_n2_words(corpus, n=None):
    vec1 = CountVectorizer(ngram_range=(2,2),
            max_features=2000).fit(corpus)
    bag_of_words = vec1.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in
                  vec1.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1],
                reverse=True)
    return words_freq[:n]
```

```python
top2_words = get_top_n2_words(data, n=200) #top 200
top2_df = pd.DataFrame(top2_words)
top2_df.columns=["Bi-gram", "Freq"]
top2_df.head()
```

```python
#Tri-gram
def get_top_n3_words(corpus, n=None):
    vec1 = CountVectorizer(ngram_range=(3,3),
           max_features=2000).fit(corpus)
    bag_of_words = vec1.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in
                  vec1.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1],
                reverse=True)
    return words_freq[:n]

top3_words = get_top_n3_words(data, n=200)
top3_df = pd.DataFrame(top3_words)
top3_df.columns=["Tri-gram", "Freq"]
```

## Naive Bayes

```python
blogs_df=pd.DataFrame({'Data':data,'Labels':blogs['Labels']})
blogs_df
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split

blog_train,blog_test = train_test_split(blogs_df,test_size=0.3)
```

```python
blog_train.shape
```

```python
def split_into_words(i):
    return (i.split(" "))
```

```python
blogs_bow = CountVectorizer(analyzer=split_into_words).fit(blogs_df.Data)
```

```python
# For training blogs
train_blogs_matrix = blogs_bow.transform(blog_train)
train_blogs_matrix.shape

# For testing blogs
test_blogs_matrix = blogs_bow.transform(blog_test)
test_blogs_matrix.shape
```

```python
from sklearn.naive_bayes import MultinomialNB as MB
from sklearn.naive_bayes import GaussianNB as GB

# Multinomial Naive Bayes
classifier_mb = MB()
classifier_mb.fit(train_blogs_matrix,blog_train.Labels)
train_pred_m = classifier_mb.predict(train_blogs_matrix)
accuracy_train_m = np.mean(train_pred_m==blog_train.Labels)
print(accuracy_train_m)

test_pred_m = classifier_mb.predict(test_blogs_matrix)
accuracy_test_m = np.mean(test_pred_m==blog_test.Labels)
print(accuracy_test_m)
```

## TFIDF

```python
# For all blogs
all_blogs_matrix = blogs_bow.transform(blogs_df.Data)
all_blogs_matrix.shape
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

# Learning Term weighting and normalizing on entire blogs
tfidf_transformer = TfidfTransformer().fit(all_blogs_matrix)

# Preparing TFIDF for train blogs
train_tfidf = tfidf_transformer.transform(train_blogs_matrix)

train_tfidf.shape

# Preparing TFIDF for test blogs
test_tfidf = tfidf_transformer.transform(test_blogs_matrix)

test_tfidf.shape
```

```python
from sklearn.naive_bayes import MultinomialNB as MB
from sklearn.naive_bayes import GaussianNB as GB

# using tfidf vectorizer
classifier_mb = MB()
classifier_mb.fit(train_tfidf,blog_train.Labels)
train_pred_m = classifier_mb.predict(train_tfidf)
accuracy_train_m = np.mean(train_pred_m==blog_train.Labels)
print(accuracy_train_m)
test_pred_m = classifier_mb.predict(test_tfidf)
accuracy_test_m = np.mean(test_pred_m==blog_test.Labels)
print(accuracy_test_m)
```

The Naive bayes model got the high accuracy value so the we can use this to predict future values.
